```python
import json
import csv
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lfile=[]
with open('data1.json',"r") as read_file:
    for line in read_file:
        line1=json.loads(line)
        lfile.append(line1['data']['Author'])
        
        
    
        
```

```python

regex1=re.compile(r'(\w)+(\s)?(\w)+')
for item in lfile:
    r=regex1.search(item)
    list1.append(r.group())
```

```python
list1
```

```python
set1=set(list1)
```

```python
for item in set1:
    dict1[item]=list1.count(item)
```

```python
list2=[]
list3=[]
for (key,value) in dict1.items():
    list2.append(key)
    list3.append(value)
    
    
```

```python
dict2={}
dict2['name']=list2
dict2['commits']=list3
```

```python
ad=pd.DataFrame.from_dict(dict2)
```

```python
top_10=ad.sort_values('commits',ascending=False).head(10)
```

```python
top_10
```

```python
plt.figure(figsize=(20,6))
x=range(10)
plt.bar(x,top_10['commits']/10)
plt.xticks(x,top_10['name'])
plt.xlabel('name')
plt.ylabel('commits')
plt.title('top 10 Author')
plt.show()
```
